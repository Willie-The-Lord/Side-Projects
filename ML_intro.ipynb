```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[-1, 2], [-4,7], [9,-8], [5, -5]])
y = np.array([7, 3, 1, 5])
```

```python
x[:]
```

```python
x[:,0]
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=y);
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x, y)
```

```python
clf.predict([[-0.8,-1]])
```

```python
gd = np.array([[i,j] for i in np.arange(-8, 4, 0.6) 
 for j in np.arange(-10, 6, 0.8)])
```

```python
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```

```python
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_features=2, n_redundant=0, 
                           n_informative=2, 
                           n_clusters_per_class=1, n_classes=3)
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

```python
clf = SVC()
clf.fit(x,y)
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x))
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x) - y)
```

```python
gd = np.array([[i,j] for i in np.arange(-4, 4, 0.4) 
 for j in np.arange(-3, 4, 0.4)])
```

```python
gdc = clf.predict(gd)
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```

```python
x = np.random.rand(100,2)
```

```python
plt.scatter(x[:,0], x[:,1], s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clf.labels_)
```

```python
gd = np.array([[i,j] for i in np.arange(-4, 4, 0.4) 
 for j in np.arange(-3, 3, 0.4)])
```

```python
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```
